$$t(x) = a\cdot sin(\pi\cdot x)$$<br>
$$a = max(f(x))$$<br>
$$c =\int_{0}^{1} t(x)dx \longrightarrow c =\int_{0}^{1} a\cdot sin(\pi\cdot x)dx\longrightarrow c = \max(f(x))\int_{0}^{1} sin(\pi\cdot x)dx$$<br>
$$c =\max(f(x))\cdot\left[-\frac{\cos(\pi \cdot x)}{\pi}\right]_0^1 = \max(f(x))\cdot \left(\frac{1}{\pi}+\frac{1}{\pi}\right) = \frac{2 \max(f(x))}{\pi}$$<br>
$$g(x) = \frac{t(x)}{c} = \frac{max(f(x))\cdot \sin(\pi \cdot x)}{\frac{2 \max(f(x))}{\pi}} = \frac{\pi\cdot \sin(\pi \cdot x) }{2}$$

**<center> Validar**<br>
$$G(x) = \int_{-\infty}^{\infty} g(x)dx=\int_{0}^{1} g(x)dx = \frac{\pi}{2}\int_{0}^{1} \sin(\pi \cdot x)dx $$
$$= -\frac{1}{2}\cdot \cos(x \pi)|_0^1 = \frac{1}{2}\cdot (1 +1) = 1$$

**Aplicamos método de transformada inversa**
$$G(x) = \int_{0}^{x} g(x)dx = \frac{\pi}{2}\int_{0}^{x} \sin(\pi \cdot x)dx $$
$$= -\frac{1}{2}\cdot \cos(x \pi)|_0^x = \frac{1}{2}\cdot (1-\cos(\pi x)) = U $$
**Despejamos**    
$$ x = \frac{\cos^{-1}(1- 2\cdot U )}{\pi} $$

```python
def Acep_rechazo(R2:'Variables distruidas U(0,1)',
                 R1:'Variables distribuidas como g(x)',
                 F:'Función objetivo a generar',
                 T:'Función que mayora a f'):
    f_x = F(R1)
    t_x = T(R1)
    condition = np.multiply(R2,t_x)<=f_x

    [plt.plot(R1[i],R2[i]*t_x[i],'ob') if condition[i] else plt.plot(R1[i],R2[i]*t_x[i],'o') for i in range(len(R1))]
                
    plt.show()
    return np.sum(condition)/len(condition),np.multiply(R2,t_x)[np.multiply(R2,t_x)<=f_x]
```

Se utiliza el método de aceptación rechazo, utilizando una funcion que sobrepase $f(x)$ siendo esta: $t(x) = a\sin(\pi x)$ entonces se generan variables aleatoria que sigan esta distribución, la cual es mas facil de integrar: 
$$\int_{0}^{x}\sin(\pi y) dy = \dfrac {1}{\pi}(-\cos(\pi x) - \dfrac {1}{\pi}) = G(X) $$
Se despeja x para con esto generar aleatorios, ya que se sabe que $G(x)$ solo va de $[0, 1]$ entonces lo podemos poner como un aleatorio uniforme entre 0 y 1
$$\dfrac{1}{\pi}\arccos(1-u\pi)  = x$$

```python
#Librerias
import numpy as np
import matplotlib.pyplot as plt
from scipy import optimize
```

```python
#Función de aceptación y rechazo usando compresión de listas
def Acep_rech(R2:'distribuida U~U(0,1)',
                 R1:'distribuida como g(x)',
                 f:'fun objetivo',
                 t:'fun que mayora a f'):
    f_x = f(R1)
    t_x = t(R1)
    condition = np.multiply(R2,t_x)<=f_x
    [plt.plot(R1[i],R2[i]*t_x[i],'ob') if condition[i] else plt.plot(R1[i],R2[i]*t_x[i],'o') \
         for i in range(len(R1))]
    elem = [x for x in range(len(R1)) if condition[x]]
    return elem            
    plt.show()
    
# Función objetivo
f = lambda x:30*(x**2-2*x**3+x**4)
# Máximo de la función f
maxi = f(optimize.fmin(lambda x:-f(x),0,disp=False))
print(maxi)
# Función t -> Función a*sin(pi*x)
x = np.arange(0, 1, 0.01)
a = maxi

t = lambda x: a*np.sin(np.pi*x)
g = lambda x: np.pi/2*np.sin(np.pi*x)

N = 10000
R2 = np.random.rand(N)
u = np.random.uniform(0, 1, N)

R1 =  (1/np.pi)*np.arccos(1-(u*np.pi))

plt.hist(R1, 30, density = True)
plt.plot(x, g(x))
plt.grid()
plt.show()

plt.plot(x, t(x), color = 'black')
plt.plot(x, f(x), label='f(x)')
plt.grid()

sim1 = Acep_rech(R2, R1, f, t)

aceptados = len(sim1)/ N
print(aceptados)
```

8.Tenemos: $$P(k + 1)=\frac{e^{-\lambda}(\lambda)^{k+1}}{(k+1)!}$$

$$P(k + 1)=\frac{e^{-\lambda}(\lambda)^{k}\lambda}{(1+k)k!}$$

Entonces:
$$P(k + 1)=\frac{\lambda}{(1+k)} P(k,\lambda) $$

```python
#8.Demostracion Poisson y generacion var aleatoria
def Poisson(k,l):
    global pr
    pr = np.zeros(k)
    pr[0] = np.exp(-l)
    def Pr(i):
        global pr
        pr[i+1]=l/(i+1)*pr[i]
        
    # Lleno el vector Pr usando compresión de listas
    [Pr(i) for i in range(k-1)]
    return np.cumsum(pr)
```

```python
N = 10000; n = 12; l = 4
u = np.random.rand(N)
p_acum = Poisson(n,l)

dist = list(map(lambda j:sum(1 for i in p_acum if i<u[j]),range(N)))

#resultados
plt.figure(figsize = [10,5])
plt.hist(dist, 100, density=True, label='Funcion')
plt.grid()
plt.legend()
```
